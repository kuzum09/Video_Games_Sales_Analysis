# Checking Skewness

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, norm
# Create a histogram of the 'User Score' data
plt.figure(figsize=(4, 3))
sns.histplot(df['User_Score'], bins=30, kde=True)
plt.xlabel('Users')
plt.ylabel('No. Of Games')
plt.title('User Score')

# Calculate skewness
skewness = skew(df['User_Score'])
print(f'Skewness: {skewness:.2f}')

# Check if the distribution is normal, right-skewed, or left-skewed
if abs(skewness) < 0.5:
    print('Distribution is approximately normal.')
elif skewness > 0.5:
    print('Distribution is right-skewed.')
else:
    print('Distribution is left-skewed.')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, norm
# Create a histogram of the 'User Score' data
plt.figure(figsize=(4, 3))
sns.histplot(df['Critic_Score'], bins=30, kde=True)
plt.xlabel('Critics')
plt.ylabel('No. Of Games')
plt.title('Critic Score')

# Calculate skewness
skewness = skew(df['Critic_Score'])
print(f'Skewness: {skewness:.2f}')

# Check if the distribution is normal, right-skewed, or left-skewed
if abs(skewness) < 0.5:
    print('Distribution is approximately normal.')
elif skewness > 0.5:
    print('Distribution is right-skewed.')
else:
    print('Distribution is left-skewed.')

plt.show()
```

From the above plot we find that the data we have is left skewed. As the plot is skewed, for further methods we use the data frame with median imputation. If the plot would have been normally distributed then we would have taken data frame with mean imputation.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import KBinsDiscretizer

# Define the number of clusters for K-Means Discretization
num_clusters = 5

# K-Means Discretization for 'Salary'
kmeans = KBinsDiscretizer(n_bins=num_clusters, encode='ordinal', strategy='kmeans')
df2['User_Score K-Means'] = kmeans.fit_transform(df2[['User_Score']])

# K-Means Discretization for 'Bonus %'
kmeans = KBinsDiscretizer(n_bins=num_clusters, encode='ordinal', strategy='kmeans')
df2['Critic_Score K-Means'] = kmeans.fit_transform(df2[['Critic_Score']])
df2['User_Score K-Means']
df2['Critic_Score K-Means']
```

```python
# Visualize the data before and after discretization for User_Score
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(df2['User_Score'], bins=30, edgecolor='k')
plt.title('Original User Score Distribution')
plt.subplot(1, 2, 2)
plt.hist(df2['User_Score K-Means'], bins=num_clusters, edgecolor='k')
plt.title('User Score K-Means Discretization')
plt.show()

# Visualize the data before and after discretization for Critic_Score
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(df2['Critic_Score'], bins=30, edgecolor='k')
plt.title('Original Critic Score Distribution')
plt.subplot(1, 2, 2)
plt.hist(df2['Critic_Score K-Means'], bins=num_clusters, edgecolor='k')
plt.title('Critic Score K-Means Discretization')
plt.show()
```

# Normalization
We are using the Z-score nomralization and Min-Max normalization techniques.
For this, "Critic_Score", "User_Score"  column are used.

<strong>Z-Score:</strong> (x_new = x_old-x_mean)/x_standard_deviation

<strong>Min-Max:</strong> (x_old-x_min)-(x_max-x_min)

```python
# Min-Max Normalization approach
df2['Critic_Score'] = (df2['Critic_Score']-df2['Critic_Score'].min())/(df2['Critic_Score'].max()-df2['Critic_Score'].min())
df2['User_Score'] = (df2['User_Score']-df2['User_Score'].min())/(df2['User_Score'].max()-df2['User_Score'].min())
```

```python
df2['Critic_Score'].head(20)
```

```python
df2['User_Score'].head(20)
```

```python
# Z-Score Normalization apporach
df2['Critic_Score'] = (df2['Critic_Score']-df2['Critic_Score'].mean())/df2['Critic_Score'].std()
df2['User_Score'] = (df2['User_Score']-df2['User_Score'].mean())/df2['User_Score'].std()
```
